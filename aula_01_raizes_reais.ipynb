```python
"""
Cálculo Numérico - Raízes Reais de Funções
==========================================
Implementação dos principais métodos numéricos para encontrar raízes de funções.
Cada método retorna (raiz, passos).
"""

import math


# =============================================================================
# 1. ERRO ABSOLUTO E RELATIVO
# =============================================================================

def erro_absoluto(valor_aproximado: float, valor_exato: float) -> float:
    """
    Calcula o erro absoluto entre um valor aproximado e o valor exato.

    Parâmetros:
        valor_aproximado: valor calculado/estimado
        valor_exato: valor de referência (verdadeiro)

    Retorna:
        Erro absoluto (float)
    """
    return abs(valor_aproximado - valor_exato)


def erro_relativo(valor_aproximado: float, valor_exato: float) -> float:
    """
    Calcula o erro relativo entre um valor aproximado e o valor exato.

    Parâmetros:
        valor_aproximado: valor calculado/estimado
        valor_exato: valor de referência (verdadeiro)

    Retorna:
        Erro relativo (float). Levanta ZeroDivisionError se valor_exato == 0.
    """
    if valor_exato == 0:
        raise ZeroDivisionError("O valor exato não pode ser zero para o cálculo do erro relativo.")
    return abs((valor_aproximado - valor_exato) / valor_exato)


def erro_relativo_percentual(valor_aproximado: float, valor_exato: float) -> float:
    """
    Calcula o erro relativo percentual.

    Retorna:
        Erro relativo em porcentagem (float).
    """
    return erro_relativo(valor_aproximado, valor_exato) * 100


# =============================================================================
# 2. MÉTODO DA BISSECÇÃO
# =============================================================================

def bisseccao(f, a: float, b: float, tol: float = 1e-6, max_iter: int = 1000):
    """
    Método da Bissecção para encontrar a raiz de f no intervalo [a, b].

    Condição: f(a) * f(b) < 0 (sinais opostos).

    Parâmetros:
        f       : função contínua f(x)
        a       : extremo esquerdo do intervalo
        b       : extremo direito do intervalo
        tol     : tolerância para o critério de parada (erro absoluto)
        max_iter: número máximo de iterações

    Retorna:
        (raiz, passos): tupla com a raiz encontrada e o número de iterações
    """
    if f(a) * f(b) >= 0:
        raise ValueError("f(a) e f(b) devem ter sinais opostos.")

    passos = 0
    c = a

    for _ in range(max_iter):
        c = (a + b) / 2.0
        passos += 1

        if abs(f(c)) < tol or (b - a) / 2.0 < tol:
            break

        if f(a) * f(c) < 0:
            b = c
        else:
            a = c

    return c, passos


# =============================================================================
# 3. MÉTODO DA POSIÇÃO FALSA (Regula Falsi)
# =============================================================================

def posicao_falsa(f, a: float, b: float, tol: float = 1e-6, max_iter: int = 1000):
    """
    Método da Posição Falsa (Regula Falsi) para encontrar a raiz de f em [a, b].

    Condição: f(a) * f(b) < 0 (sinais opostos).

    Parâmetros:
        f       : função contínua f(x)
        a       : extremo esquerdo do intervalo
        b       : extremo direito do intervalo
        tol     : tolerância para o critério de parada
        max_iter: número máximo de iterações

    Retorna:
        (raiz, passos): tupla com a raiz encontrada e o número de iterações
    """
    if f(a) * f(b) >= 0:
        raise ValueError("f(a) e f(b) devem ter sinais opostos.")

    passos = 0
    c = a

    for _ in range(max_iter):
        fa, fb = f(a), f(b)
        c = (a * fb - b * fa) / (fb - fa)
        fc = f(c)
        passos += 1

        if abs(fc) < tol:
            break

        if fa * fc < 0:
            b = c
        else:
            a = c

    return c, passos


# =============================================================================
# 4. MÉTODO ITERATIVO LINEAR (Ponto Fixo)
# =============================================================================

def iterativo_linear(g, x0: float, tol: float = 1e-6, max_iter: int = 1000):
    """
    Método Iterativo Linear (Ponto Fixo) para encontrar a raiz de f(x) = 0,
    reescrita como x = g(x).

    Parâmetros:
        g       : função de iteração g(x), tal que a raiz de f é ponto fixo de g
        x0      : estimativa inicial
        tol     : tolerância para o critério de parada (erro absoluto entre iterações)
        max_iter: número máximo de iterações

    Retorna:
        (raiz, passos): tupla com a raiz encontrada e o número de iterações

    Observação:
        O método converge se |g'(x)| < 1 na vizinhança da raiz.
    """
    x_atual = x0
    passos = 0

    for _ in range(max_iter):
        x_novo = g(x_atual)
        passos += 1

        if abs(x_novo - x_atual) < tol:
            return x_novo, passos

        x_atual = x_novo

    return x_atual, passos


# =============================================================================
# 5. MÉTODO DE NEWTON-RAPHSON
# =============================================================================

def newton_raphson(f, df, x0: float, tol: float = 1e-6, max_iter: int = 1000):
    """
    Método de Newton-Raphson para encontrar a raiz de f(x) = 0.

    Parâmetros:
        f       : função f(x)
        df      : derivada de f, isto é, f'(x)
        x0      : estimativa inicial
        tol     : tolerância para o critério de parada
        max_iter: número máximo de iterações

    Retorna:
        (raiz, passos): tupla com a raiz encontrada e o número de iterações

    Levanta:
        ZeroDivisionError se a derivada for zero em algum ponto.
    """
    x = x0
    passos = 0

    for _ in range(max_iter):
        fx = f(x)
        dfx = df(x)
        passos += 1

        if abs(dfx) < 1e-14:
            raise ZeroDivisionError(f"Derivada nula em x = {x}. Método não pode continuar.")

        x_novo = x - fx / dfx

        if abs(x_novo - x) < tol:
            return x_novo, passos

        x = x_novo

    return x, passos


# =============================================================================
# 6. MÉTODO DA SECANTE
# =============================================================================

def secante(f, x0: float, x1: float, tol: float = 1e-6, max_iter: int = 1000):
    """
    Método da Secante para encontrar a raiz de f(x) = 0.

    Não exige o cálculo explícito da derivada; aproxima-a por diferença finita.

    Parâmetros:
        f       : função f(x)
        x0      : primeira estimativa inicial
        x1      : segunda estimativa inicial
        tol     : tolerância para o critério de parada
        max_iter: número máximo de iterações

    Retorna:
        (raiz, passos): tupla com a raiz encontrada e o número de iterações

    Levanta:
        ZeroDivisionError se f(x1) - f(x0) == 0 durante a iteração.
    """
    passos = 0

    for _ in range(max_iter):
        f0, f1 = f(x0), f(x1)
        passos += 1

        if abs(f1 - f0) < 1e-14:
            raise ZeroDivisionError("Divisão por zero no Método da Secante: f(x1) ≈ f(x0).")

        x2 = x1 - f1 * (x1 - x0) / (f1 - f0)

        if abs(x2 - x1) < tol:
            return x2, passos

        x0, x1 = x1, x2

    return x1, passos


# =============================================================================
# DEMONSTRAÇÃO / TESTES
# =============================================================================

if __name__ == "__main__":
    print("=" * 60)
    print("  CÁLCULO NUMÉRICO — Raízes Reais de Funções")
    print("=" * 60)

    # Função de teste: f(x) = x^3 - x - 2  →  raiz ≈ 1.5213797
    def f(x):
        return 3*x**2 - 27  # x**3 - x - 2

    def df(x):
        return 3 * x**2 - 1

    def g(x):           # x = g(x) para ponto fixo: x = (x + 2)^(1/3)
        return (x + 2) ** (1 / 3)

    raiz_exata = 1.5213797068045676

    print(f"\nFunção: f(x) = x³ - x - 2")
    print(f"Raiz exata (referência): {raiz_exata:.10f}\n")

    # --- Erros ---
    aprox = 1.52
    print("1. Erro Absoluto e Relativo")
    print(f"   Valor aproximado : {aprox}")
    print(f"   Erro absoluto    : {erro_absoluto(aprox, raiz_exata):.2e}")
    print(f"   Erro relativo    : {erro_relativo(aprox, raiz_exata):.2e}")
    print(f"   Erro relativo %  : {erro_relativo_percentual(aprox, raiz_exata):.4f}%\n")

    # --- Bissecção ---
    raiz, passos = bisseccao(f, -4, -1, 1e-2, 7)
    print(f"2. Bissecção")
    print(f"   Raiz   : {raiz:.10f}  |  Passos: {passos}\n")

    # --- Posição Falsa ---
    raiz, passos = posicao_falsa(f, -4, -1, 1e-2, 7)
    print(f"3. Posição Falsa")
    print(f"   Raiz   : {raiz:.10f}  |  Passos: {passos}\n")

    # --- Iterativo Linear ---
    raiz, passos = iterativo_linear(g, x0=1.5)
    print(f"4. Iterativo Linear (Ponto Fixo)")
    print(f"   Raiz   : {raiz:.10f}  |  Passos: {passos}\n")

    # --- Newton-Raphson ---
    raiz, passos = newton_raphson(f, df, x0=1.5)
    print(f"5. Newton-Raphson")
    print(f"   Raiz   : {raiz:.10f}  |  Passos: {passos}\n")

    # --- Secante ---
    raiz, passos = secante(f, x0=1.0, x1=2.0)
    print(f"6. Secante")
    print(f"   Raiz   : {raiz:.10f}  |  Passos: {passos}\n")

    print("=" * 60)
```
