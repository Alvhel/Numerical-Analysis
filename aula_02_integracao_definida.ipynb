```python
# codigo original

import math

import sympy as sp


def generate_points(a: float, b: float, n: int):
    x = [a + i * (b - a) / n for i in range(n + 1)]
    h = (b - a) / n
    return x, h


def integral(
    a: int, b: int
):  # otimizar a função que recebe e calcula a integral -> usar OOP?
    x = sp.symbols("x")

    f = 2,71828182845904**x**2  #-(x ** 5) + (3 * x ** 3) + x ** 2  # sp.exp(-2*x) + x**2

    integral = sp.integrate(f, (x, a, b))

    return float(integral)


def compute_fx(x):  # otimizar a função que calcula Fx -> usar OOP?
    return  [math.exp(2,71828182845904**x**2)] # [-(i ** 5) + (3 * i ** 3) + i ** 2 for i in x]
    # [math.exp(-2 * i) + i ** 2 for i in x]


def simpson_rules(fx: list, n: int, h: float):
    n = len(fx) - 1
    first_last = fx[0] + fx[-1]

    if n % 2 == 0:  # Simpson 1/3

        odds = fx[1:-1:2]
        evens = fx[2:-1:2]

        simpson_1_3 = (h / 3) * (first_last + 4 * sum(odds) + 2 * sum(evens))

        return simpson_1_3

    elif n % 3 == 0:  # Simpson 3/8

        multiple_3 = []
        multiple_any = []

        for i in range(1, n):

            if i % 3 == 0:
                multiple_3.append(fx[i])
            else:
                multiple_any.append(fx[i])

        simpson_3_8 = (3 * h / 8) * (
            first_last + 3 * sum(multiple_any) + 2 * sum(multiple_3)
        )

        return simpson_3_8

    else:
        return (
            "Simpson Methods are not available in this scenario, try Trapezius Method."
        )


def trapezius_method(fx: list, n: int, h: float):
    first_last = fx[0] + fx[-1]
    return h / 2 * (first_last + 2 * sum(fx[1:-1]))


def rectangle_by_right(fx: list, n: int, h: float):
    return h * sum(fx[1:])


def rectangle_by_left(fx: list, n: int, h: float):
    return h * sum(fx[0:-1])


def rectangle_by_center(x: list, n: int, h: float):
    xm = [((x[i] + x[i + 1]) / 2) for i in range(len(x) - 1)]
    fxm = compute_fx(xm)
    print("\n" + 20 * "=" + " Metodo dos Retangulos com Altura Centrada " + 20 * "=")
    print(f"\nXm = {xm}\n")
    print(f"Fxm = {fxm}\n")
    return h * sum(fxm[:])


def main():

    a = float(input("Valor de a: "))
    b = float(input("Valor de b: "))
    n = int(input("Valor de subintervalos (n): "))

    x, h = generate_points(a, b, n)
    fx = compute_fx(x)
    integral_exata = integral(a, b)

    simpson = simpson_rules(fx, n, h)
    trapezius = trapezius_method(fx, n, h)
    left_rectangle = rectangle_by_left(fx, n, h)
    right_rectangle = rectangle_by_right(fx, n, h)
    center_rectangle = rectangle_by_center(x, n, h)

    aprox = [simpson, trapezius, left_rectangle, right_rectangle, center_rectangle]
    melhor = min(aprox, key=lambda x: abs(x - integral_exata))

    print(f"\nIntegral exata: {integral_exata}\n")
    print(f"x = {x}\n")
    print(f"F(x) = {fx}\n")
    print(f"Por Simpson = {simpson}")
    print(f"Por Trapézio = {trapezius}")
    print(f"Por Retângulo tomada pela esquerda = {left_rectangle}")
    print(f"Por Retângulo tomada pela direita = {right_rectangle}")
    print(f"Por Retângulo tomada pelo centro = {center_rectangle}")

    print(f"\nMenor Erro Absoluto {abs(integral_exata - melhor)}")
    print(f"Menor Erro Relativo {abs(integral_exata - melhor) / abs(integral_exata)}")


main()
```

```python
# INTEGRAIS DEFINIDAS, REFATORADO.
import math

import sympy as sp


# cria função a partir do input do usuário
def build_function(expr_str):

    expr_str = expr_str.replace("^", "**")

    x = sp.symbols("x")

    expr = sp.sympify(expr_str)  # expressão simbólica
    f = sp.lambdify(x, expr, "math")  # função numérica

    return f, expr


def exact_integral(expr, a, b):

    x = sp.symbols("x")

    integral = sp.integrate(expr, (x, a, b))

    return float(integral)


def generate_x(a, b, n):

    h = (b - a) / n

    x = [a + i * h for i in range(n + 1)]

    return x, h


def simpson_1_3(f, a, b, n):

    x, h = generate_x(a, b, n)

    fx = [f(i) for i in x]

    return (h / 3) * (fx[0] + fx[-1] + 4 * sum(fx[1:-1:2]) + 2 * sum(fx[2:-1:2]))


def simpson_3_8(f, a, b, n):

    x, h = generate_x(a, b, n)

    fx = [f(i) for i in x]

    return (3 * h / 8) * (
        fx[0]
        + fx[-1]
        + 3 * sum(fx[i] for i in range(1, n) if i % 3 != 0)
        + 2 * sum(fx[i] for i in range(3, n, 3))
    )


def trapezoid(f, a, b, n):

    x, h = generate_x(a, b, n)

    fx = [f(i) for i in x]

    return (h / 2) * (fx[0] + fx[-1] + 2 * sum(fx[1:-1]))


def rectangle_left(f, a, b, n):

    x, h = generate_x(a, b, n)

    fx = [f(i) for i in x]

    return h * sum(fx[:-1])


def rectangle_right(f, a, b, n):

    x, h = generate_x(a, b, n)

    fx = [f(i) for i in x]

    return h * sum(fx[1:])


def rectangle_center(f, a, b, n):

    h = (b - a) / n

    xm = [a + (i + 0.5) * h for i in range(n)]

    return h * sum(f(i) for i in xm)


def main():

    print("\n===== Integração Numérica =====\n")

    expr_str = input("Digite a função f(x): ")

    f, expr = build_function(expr_str)

    a = float(input("Valor de a: "))
    b = float(input("Valor de b: "))
    n = int(input("Valor de subintervalos (n): "))

    exact = exact_integral(expr, a, b)

    results = {
        "Simpson 1/3": simpson_1_3(f, a, b, n) if n % 2 == 0 else None,
        "Simpson 3/8": simpson_3_8(f, a, b, n) if n % 3 == 0 else None,
        "Trapézio": trapezoid(f, a, b, n),
        "Retângulo Esquerda": rectangle_left(f, a, b, n),
        "Retângulo Direita": rectangle_right(f, a, b, n),
        "Retângulo Centro": rectangle_center(f, a, b, n),
    }

    print(f"\nIntegral exata = {exact}\n")

    valid_results = []

    for name, value in results.items():

        if value is not None:

            erro_abs = abs(exact - value)
            erro_rel = erro_abs / abs(exact)

            print(f"{name}: {value}")
            print(f"Erro absoluto: {erro_abs}")
            print(f"Erro relativo: {erro_rel}\n")

            valid_results.append((name, erro_abs))

        else:

            print(f"{name}: método não aplicável para n = {n}\n")

    best = min(valid_results, key=lambda x: x[1])

    print("===== Melhor Aproximação =====")
    print(f"Método: {best[0]}")
    print(f"Menor erro absoluto: {best[1]}")


main()
```

```python
# Integração Numérica - Versão Otimizada Claude
"""
Integração Numérica - Versão Otimizada
======================================
Métodos implementados:
  - Simpson 1/3
  - Simpson 3/8
  - Trapézio
  - Retângulo (Esquerda, Direita, Centro)

Melhorias em relação à versão original:
  - Uso de numpy para vetorização (substitui list comprehensions manuais)
  - Separação clara de responsabilidades (sem estado global)
  - Tipagem explícita com type hints
  - Docstrings descritivas
  - Tratamento de erros com exceções adequadas
  - Nenhuma duplicação de generate_x entre métodos
  - Impressão organizada via dataclass de resultado
"""

from __future__ import annotations

from dataclasses import dataclass, field
from typing import Callable

import numpy as np
import sympy as sp


# ---------------------------------------------------------------------------
# Tipos auxiliares
# ---------------------------------------------------------------------------

FloatFunc = Callable[[float], float]


# ---------------------------------------------------------------------------
# Dataclass de resultado — evita dicionário solto no main()
# ---------------------------------------------------------------------------

@dataclass
class IntegrationResult:
    method: str
    value: float
    abs_error: float
    rel_error: float

    def __str__(self) -> str:
        return (
            f"{self.method}: {self.value:.10f}\n"
            f"  Erro absoluto : {self.abs_error:.2e}\n"
            f"  Erro relativo : {self.rel_error:.2e}"
        )


# ---------------------------------------------------------------------------
# Construção da função a partir de string
# ---------------------------------------------------------------------------

def build_function(expr_str: str) -> tuple[FloatFunc, sp.Expr]:
    """
    Converte uma string matemática em uma função numérica e uma expressão
    simbólica.

    Aceita notação '^' como alternativa a '**'.

    Nota: quando a expressão não contém 'x' (constante), lambdify retorna
    um escalar. O wrapper garante que a saída seja sempre um np.ndarray.
    """
    expr_str = expr_str.replace("^", "**")
    x = sp.Symbol("x")
    expr = sp.sympify(expr_str)
    _f = sp.lambdify(x, expr, modules="numpy")

    def f(x_val):
        result = _f(x_val)
        x_arr = np.asarray(x_val)
        return np.broadcast_to(np.asarray(result, dtype=float), x_arr.shape).copy()

    return f, expr


# ---------------------------------------------------------------------------
# Integral exata (simbólica)
# ---------------------------------------------------------------------------

def exact_integral(expr: sp.Expr, a: float, b: float) -> float:
    """Calcula a integral definida de forma exata usando SymPy."""
    x = sp.Symbol("x")
    result = sp.integrate(expr, (x, a, b))
    return float(result)


# ---------------------------------------------------------------------------
# Geração de pontos — centralizada, chamada uma única vez por método
# ---------------------------------------------------------------------------

def _linspace(a: float, b: float, n: int) -> tuple[np.ndarray, float]:
    """
    Retorna n+1 pontos igualmente espaçados em [a, b] e o passo h.
    Usa np.linspace para evitar acúmulo de erro de ponto flutuante
    que ocorre na soma iterativa  a + i * h.
    """
    x = np.linspace(a, b, n + 1)
    h = (b - a) / n
    return x, h


# ---------------------------------------------------------------------------
# Métodos de integração numérica
# ---------------------------------------------------------------------------

def simpson_1_3(f: FloatFunc, a: float, b: float, n: int) -> float | None:
    """
    Regra de Simpson 1/3.
    Requer n par. Retorna None caso contrário.
    """
    if n % 2 != 0:
        return None

    x, h = _linspace(a, b, n)
    fx = f(x)  # vetorizado via numpy

    # Coeficientes: 1, 4, 2, 4, 2, ..., 4, 1
    coefs = np.ones(n + 1)
    coefs[1:-1:2] = 4   # índices ímpares
    coefs[2:-2:2] = 2   # índices pares internos

    return (h / 3) * np.dot(coefs, fx)


def simpson_3_8(f: FloatFunc, a: float, b: float, n: int) -> float | None:
    """
    Regra de Simpson 3/8.
    Requer n múltiplo de 3. Retorna None caso contrário.
    """
    if n % 3 != 0:
        return None

    x, h = _linspace(a, b, n)
    fx = f(x)

    # Coeficientes: 1, 3, 3, 2, 3, 3, 2, ..., 3, 3, 1
    coefs = np.ones(n + 1)
    for i in range(1, n):
        coefs[i] = 2 if i % 3 == 0 else 3

    return (3 * h / 8) * np.dot(coefs, fx)


def trapezoid(f: FloatFunc, a: float, b: float, n: int) -> float:
    """Método do Trapézio composto."""
    x, h = _linspace(a, b, n)
    fx = f(x)

    # np.trapezoid (numpy >= 2.0) ou np.trapz (numpy < 2.0)
    trapz = getattr(np, "trapezoid", getattr(np, "trapz", None))
    return float(trapz(fx, x))


def rectangle_left(f: FloatFunc, a: float, b: float, n: int) -> float:
    """Retângulos tomados pelo extremo esquerdo de cada subintervalo."""
    x, h = _linspace(a, b, n)
    return h * np.sum(f(x[:-1]))


def rectangle_right(f: FloatFunc, a: float, b: float, n: int) -> float:
    """Retângulos tomados pelo extremo direito de cada subintervalo."""
    x, h = _linspace(a, b, n)
    return h * np.sum(f(x[1:]))


def rectangle_center(f: FloatFunc, a: float, b: float, n: int) -> float:
    """Retângulos com altura no ponto médio de cada subintervalo."""
    h = (b - a) / n
    # Pontos médios: a + (i + 0.5)*h  para  i = 0..n-1
    xm = np.linspace(a + h / 2, b - h / 2, n)
    return h * np.sum(f(xm))


# ---------------------------------------------------------------------------
# Executor — roda todos os métodos e coleta resultados
# ---------------------------------------------------------------------------

def run_all(
    f: FloatFunc,
    expr: sp.Expr,
    a: float,
    b: float,
    n: int,
) -> tuple[float, list[IntegrationResult]]:
    """
    Executa todos os métodos de integração e devolve a integral exata
    junto com a lista de resultados.
    """
    exact = exact_integral(expr, a, b)

    methods: dict[str, float | None] = {
        "Simpson 1/3":         simpson_1_3(f, a, b, n),
        "Simpson 3/8":         simpson_3_8(f, a, b, n),
        "Trapézio":            trapezoid(f, a, b, n),
        "Retângulo Esquerda":  rectangle_left(f, a, b, n),
        "Retângulo Direita":   rectangle_right(f, a, b, n),
        "Retângulo Centro":    rectangle_center(f, a, b, n),
    }

    results: list[IntegrationResult] = []
    for name, value in methods.items():
        if value is not None:
            abs_err = abs(exact - value)
            rel_err = abs_err / abs(exact) if exact != 0 else float("inf")
            results.append(IntegrationResult(name, value, abs_err, rel_err))

    return exact, results


# ---------------------------------------------------------------------------
# Interface de linha de comando
# ---------------------------------------------------------------------------

def main() -> None:
    print("\n===== Integração Numérica =====\n")

    expr_str = input("Digite a função f(x): ").strip()
    a = float(input("Valor de a: "))
    b = float(input("Valor de b: "))
    n = int(input("Valor de n (subintervalos): "))

    if n <= 0:
        raise ValueError("n deve ser um inteiro positivo.")
    if a >= b:
        raise ValueError("O limite inferior a deve ser menor que b.")

    try:
        f, expr = build_function(expr_str)
    except Exception as exc:
        raise ValueError(f"Expressão inválida: '{expr_str}'") from exc

    exact, results = run_all(f, expr, a, b, n)

    print(f"\nIntegral exata = {exact:.10f}\n")
    print("-" * 50)

    for r in results:
        print(r)
        print()

    # Métodos não aplicáveis
    if n % 2 != 0:
        print("Simpson 1/3 : não aplicável (n deve ser par)\n")
    if n % 3 != 0:
        print("Simpson 3/8 : não aplicável (n deve ser múltiplo de 3)\n")

    if results:
        best = min(results, key=lambda r: r.abs_error)
        print("=" * 50)
        print(f"Melhor aproximação : {best.method}")
        print(f"Erro absoluto      : {best.abs_error:.2e}")
        print(f"Erro relativo      : {best.rel_error:.2e}")


if __name__ == "__main__":
    main()
```
