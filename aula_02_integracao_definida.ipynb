```python
import math

import sympy as sp


def generate_points(a: float, b: float, n: int):
    x = [a + i * (b - a) / n for i in range(n + 1)]
    h = (b - a) / n
    return x, h


def integral(
    a: int, b: int
):  # otimizar a função que recebe e calcula a integral -> usar OOP?
    x = sp.symbols("x")

    f = -(x ** 5) + (3 * x ** 3) + x ** 2  # sp.exp(-2*x) + x**2

    integral = sp.integrate(f, (x, a, b))

    return float(integral)


def compute_fx(x):  # otimizar a função que calcula Fx -> usar OOP?
    return [-(i ** 5) + (3 * i ** 3) + i ** 2 for i in x]
    # [math.exp(-2 * i) + i ** 2 for i in x]


def simpson_rules(fx: list, n: int, h: float):
    n = len(fx) - 1
    first_last = fx[0] + fx[-1]

    if n % 2 == 0:  # Simpson 1/3

        odds = fx[1:-1:2]
        evens = fx[2:-1:2]

        simpson_1_3 = (h / 3) * (first_last + 4 * sum(odds) + 2 * sum(evens))

        return simpson_1_3

    elif n % 3 == 0:  # Simpson 3/8

        multiple_3 = []
        multiple_any = []

        for i in range(1, n):

            if i % 3 == 0:
                multiple_3.append(fx[i])
            else:
                multiple_any.append(fx[i])

        simpson_3_8 = (3 * h / 8) * (
            first_last + 3 * sum(multiple_any) + 2 * sum(multiple_3)
        )

        return simpson_3_8

    else:
        return (
            "Simpson Methods are not available in this scenario, try Trapezius Method."
        )


def trapezius_method(fx: list, n: int, h: float):
    first_last = fx[0] + fx[-1]
    return h / 2 * (first_last + 2 * sum(fx[1:-1]))


def rectangle_by_right(fx: list, n: int, h: float):
    return h * sum(fx[1:])


def rectangle_by_left(fx: list, n: int, h: float):
    return h * sum(fx[0:-1])


def rectangle_by_center(x: list, n: int, h: float):
    xm = [((x[i] + x[i + 1]) / 2) for i in range(len(x) - 1)]
    fxm = compute_fx(xm)
    print("\n" + 20 * "=" + " Metodo dos Retangulos com Altura Centrada " + 20 * "=")
    print(f"\nXm = {xm}\n")
    print(f"Fxm = {fxm}\n")
    return h * sum(fxm[:])


def main():

    a = float(input("Valor de a: "))
    b = float(input("Valor de b: "))
    n = int(input("Valor de subintervalos (n): "))

    x, h = generate_points(a, b, n)
    fx = compute_fx(x)
    integral_exata = integral(a, b)

    simpson = simpson_rules(fx, n, h)
    trapezius = trapezius_method(fx, n, h)
    left_rectangle = rectangle_by_left(fx, n, h)
    right_rectangle = rectangle_by_right(fx, n, h)
    center_rectangle = rectangle_by_center(x, n, h)

    aprox = [simpson, trapezius, left_rectangle, right_rectangle, center_rectangle]
    melhor = min(aprox, key=lambda x: abs(x - integral_exata))

    print(f"\nIntegral exata: {integral_exata}\n")
    print(f"x = {x}\n")
    print(f"F(x) = {fx}\n")
    print(f"Por Simpson = {simpson}")
    print(f"Por Trapézio = {trapezius}")
    print(f"Por Retângulo tomada pela esquerda = {left_rectangle}")
    print(f"Por Retângulo tomada pela direita = {right_rectangle}")
    print(f"Por Retângulo tomada pelo centro = {center_rectangle}")

    print(f"\nMenor Erro Absoluto {abs(integral_exata - melhor)}")
    print(f"Menor Erro Relativo {abs(integral_exata - melhor) / abs(integral_exata)}")


main()
```

```python
# INTEGRAIS DEFINIDAS, REFATORADO.
import math

import sympy as sp


# cria função a partir do input do usuário
def build_function(expr_str):

    expr_str = expr_str.replace("^", "**")

    x = sp.symbols("x")

    expr = sp.sympify(expr_str)  # expressão simbólica
    f = sp.lambdify(x, expr, "math")  # função numérica

    return f, expr


def exact_integral(expr, a, b):

    x = sp.symbols("x")

    integral = sp.integrate(expr, (x, a, b))

    return float(integral)


def generate_x(a, b, n):

    h = (b - a) / n

    x = [a + i * h for i in range(n + 1)]

    return x, h


def simpson_1_3(f, a, b, n):

    x, h = generate_x(a, b, n)

    fx = [f(i) for i in x]

    return (h / 3) * (fx[0] + fx[-1] + 4 * sum(fx[1:-1:2]) + 2 * sum(fx[2:-1:2]))


def simpson_3_8(f, a, b, n):

    x, h = generate_x(a, b, n)

    fx = [f(i) for i in x]

    return (3 * h / 8) * (
        fx[0]
        + fx[-1]
        + 3 * sum(fx[i] for i in range(1, n) if i % 3 != 0)
        + 2 * sum(fx[i] for i in range(3, n, 3))
    )


def trapezoid(f, a, b, n):

    x, h = generate_x(a, b, n)

    fx = [f(i) for i in x]

    return (h / 2) * (fx[0] + fx[-1] + 2 * sum(fx[1:-1]))


def rectangle_left(f, a, b, n):

    x, h = generate_x(a, b, n)

    fx = [f(i) for i in x]

    return h * sum(fx[:-1])


def rectangle_right(f, a, b, n):

    x, h = generate_x(a, b, n)

    fx = [f(i) for i in x]

    return h * sum(fx[1:])


def rectangle_center(f, a, b, n):

    h = (b - a) / n

    xm = [a + (i + 0.5) * h for i in range(n)]

    return h * sum(f(i) for i in xm)


def main():

    print("\n===== Integração Numérica =====\n")

    expr_str = input("Digite a função f(x): ")

    f, expr = build_function(expr_str)

    a = float(input("Valor de a: "))
    b = float(input("Valor de b: "))
    n = int(input("Valor de subintervalos (n): "))

    exact = exact_integral(expr, a, b)

    results = {
        "Simpson 1/3": simpson_1_3(f, a, b, n) if n % 2 == 0 else None,
        "Simpson 3/8": simpson_3_8(f, a, b, n) if n % 3 == 0 else None,
        "Trapézio": trapezoid(f, a, b, n),
        "Retângulo Esquerda": rectangle_left(f, a, b, n),
        "Retângulo Direita": rectangle_right(f, a, b, n),
        "Retângulo Centro": rectangle_center(f, a, b, n),
    }

    print(f"\nIntegral exata = {exact}\n")

    valid_results = []

    for name, value in results.items():

        if value is not None:

            erro_abs = abs(exact - value)
            erro_rel = erro_abs / abs(exact)

            print(f"{name}: {value}")
            print(f"Erro absoluto: {erro_abs}")
            print(f"Erro relativo: {erro_rel}\n")

            valid_results.append((name, erro_abs))

        else:

            print(f"{name}: método não aplicável para n = {n}\n")

    best = min(valid_results, key=lambda x: x[1])

    print("===== Melhor Aproximação =====")
    print(f"Método: {best[0]}")
    print(f"Menor erro absoluto: {best[1]}")


main()
```
