```python
# Regra de Cramer. A = 3x3
# Matriz quadrática
# Determinante != 0
"""
    O padrão de arquivo txt deve ser somente com números e espacos, sendo pertencentes
    a B os elementos da última coluna da direita para a esquerda:
    A = Matriz
    B = Termos Independentes
    [A 3x3][B 3x1]:
        a11 a12 a13 b11
        a21 a22 a23 b21
        a31 a32 a33 b31

    """
from math import sqrt


def matrix_file(file):
    a = []
    b = []

    with open(file, "r", encoding="utf-8") as f:

        for line in f:
            values = [float(x) for x in line.split()]
            if values:
                a.append(values[0:-1])
                b.append(values[-1])

    return [a, b]


def is_square(n):
    if n < 0:
        return False
    root = sqrt(n)
    return root.is_integer()


def determinant_sarrus(matrix):
    sarrus = []
    for line in matrix:
        new_line = line + line[0:2]
        sarrus.append(new_line)

    positive = sum(sarrus[0][i] * sarrus[1][i + 1] * sarrus[2][i + 2] for i in range(3))
    negative = sum(sarrus[0][i + 2] * sarrus[1][i + 1] * sarrus[2][i] for i in range(3))

    det = positive - negative

    return det


def cramer(matrix, terms, column_index):  # detAi / detA
    new_matrix = [line[:] for line in matrix]

    for i in range(len(new_matrix)):
        new_matrix[i][column_index] = terms[i]

    return new_matrix


def main():
    path = "aula_03_matriz.txt"  # considerando que o txt esteja no mesmo diretorio do script
    A, b = matrix_file(path)

    lines = len(A)
    cols = len(A[0]) if A else 0
    elements = is_square(16)
    det = determinant_sarrus(A)

    print(f"Matriz A: {A}")
    print(f"Termos independentes B: {b}")
    print(f"Determinante de A: {det}")

    if elements == True and det != 0:
        A1 = cramer(A, b, 0)
        detA1 = determinant_sarrus(A1)

        A2 = cramer(A, b, 1)
        detA2 = determinant_sarrus(A2)

        A3 = cramer(A, b, 2)
        detA3 = determinant_sarrus(A3)

        print(f"Solucao do sistema por Cramer: {(detA1/det), (detA2/det), (detA3/det)}")

    else:
        print("A regra de Cramer não pode ser usada")


main()
```
